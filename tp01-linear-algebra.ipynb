## Exercise b)  Vector Operations

### Perform the following vector operations when possible and visualise each time the original vectors as well as how we manipulate them in order to obtain the final result, use grids such as the following for your visualisations.

```python
# Compute the result of the operation on the two vectors
def plot_vectors_operations(a, b, operation):
    
    if np.shape(a) != np.shape(b):
        raise ValueError('Vectors of different dimension')
    
    # Create a figure and axis object
    fig, ax = plt.subplots()

    # Plot the first vector in blue
    ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a')
    
    # do the maths and plot the second vector in red following the head to tails methods for sum/substraction
    if operation == '+':
        c = a + b
        ax.quiver(a[0], a[1], b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label='b')
    elif operation == '-':
        c = a - b
        ax.quiver(a[0], a[1], -b[0], -b[1], angles='xy', scale_units='xy', scale=1, color='green', label='b')
    elif operation == '*':
        c = a * b
        ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label='b')
    elif operation == '/':
        c = a / b
        ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='green', label='b')
    else:
        raise ValueError('Invalid operation: {}'.format(operation))
    
    # Plot the result of the operation in red
    ax.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='red', label='{}(a, b)'.format(operation))
    
    # Calculate the limits of the x and y axes
    minx = min(0, a[0], b[0], c[0]) - 1
    maxx = max(0, a[0], b[0], c[0]) + 1 
    ax.set_xticks(range(minx,maxx))
    ax.set_xlim([minx,maxx])
    miny = min(0, a[1], b[1], c[1]) - 1
    maxy = max(0, a[1], b[1], c[1]) + 1
    ax.set_yticks(range(miny,maxy))
    ax.set_ylim([miny,maxy])

    
    # Make the spines meet at (0,0)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Add labels for the x and y axes
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    # Add a legend
    ax.legend()

    # return the plot
    return plt
```

### Explain what happens to a vector $\textbf{a} \in \mathbb{R}^d$ when we multiply it with a scalar $\textbf{b} \in \mathbb{R}$, by exploring the following cases: $|b| < 1$, $|b| > 1$

For $\textbf{a} = \begin{bmatrix} \alpha_{1} \\ ... \\ \alpha_{d} \end{bmatrix}$ and any $b$ we have $b \times \textbf{a} = \begin{bmatrix} b \times \alpha_{1} \\ ... \\ b \times \alpha_{d} \end{bmatrix}$

If $|b| < 1$ then $ b = 1 - b'$ so

&nbsp;&nbsp;&nbsp;&nbsp;$b\textbf{a} = (1 - b')\textbf{a} = \textbf{a} - b'\textbf{a} = \begin{bmatrix} (1 - b') \times \alpha_{1} \\ ... \\ (1 - b') \times \alpha_{d} \end{bmatrix}$ 

So the multiplication of a vector by a scalar less than 1 will result in a vector $\textbf{a}$ reduce by a factor $b'$ which mean that the new vectors will be *shorter* than the original $\textbf{a}$ vectors

You can observe this with the calculus of the $\textbf{f}$ previously.

If $|b| > 1$ then $ b = 1 + b'$ so

&nbsp;&nbsp;&nbsp;&nbsp;$b\textbf{a} = (1 + b')\textbf{a} = \textbf{a} + b'\textbf{a} = \begin{bmatrix} (1 + b') \times \alpha_{1} \\ ... \\ (1 + b') \times \alpha_{d} \end{bmatrix}$ 

So the multiplication of a vector by a scalar greater than 1 will result in a vector $\textbf{a}$ increase by a factor $b'$ which mean that the new vectors will be *longer* than the original $\textbf{a}$ vectors

You can observe this with the calculus of the $\textbf{e}$ previously.

## Exercice c) Vector Operation Properties

###  Demonstrate that the following properties hold for any vectors $\textbf{a}, \textbf{b}, \textbf{c} \in \mathbb{R}^d$ and scalars $x, y, z \in \mathbb{R}$ :

#### Vector addition is commutative

The commutative property of vectors addition is verified if $\textbf{a}+\textbf{b}=\textbf{b}+\textbf{a}$

Let's prove it !

$\textbf{a}+\textbf{b}=\begin{bmatrix} a_{1} \\ ... \\ a_{d} \end{bmatrix} + \begin{bmatrix} b_{1} \\ ... \\ b_{d} \end{bmatrix} = \begin{bmatrix} a_{1} +  b_{1} \\ ... \\ a_{d} +  b_{d} \end{bmatrix} = \begin{bmatrix} b_{1} +  a_{1} \\ ... \\ b_{d} +  a_{d} \end{bmatrix} = \begin{bmatrix} b_{1} \\ ... \\ b_{d} \end{bmatrix} + \begin{bmatrix} a_{1} \\ ... \\ a_{d} \end{bmatrix} =\textbf{b}+\textbf{a} $

We prove the equality is correct, therefore we can confirm that **vector addition is commutative**.

#### Scalar multiplication is associative

The commutative property of scalar and vectors multiplication is defined by

$x \textbf{a} = \textbf{a} x$

From this affirmation we can make the following hypothesis

$(xy)\textbf{a} = x(y\textbf{a})$

Let's prove it !

On the left side, we got first a scalar-scalar multiplication and a scalar-vector multiplication

$ (xy)\textbf{a} = (xy)\begin{bmatrix} a_{1} \\ ... \\ a_{d} \end{bmatrix} = \begin{bmatrix} xya_{1} \\ ... \\ xya_{d} \end{bmatrix}$

On the right side, we got two scalar-vector multiplication

$ x(y\textbf{a}) = x(y\begin{bmatrix} a_{1} \\ ... \\ a_{d} \end{bmatrix}) = x\begin{bmatrix} ya_{1} \\ ... \\ ya_{d} \end{bmatrix} = \begin{bmatrix} xya_{1} \\ ... \\ xya_{d} \end{bmatrix}$

We prove the equality is correct, therefore we can write the vector above as $xy\textbf{a}$, since it does not matter whether
we interpret this as $x(y\textbf{a})$ or $(xy)\textbf{a}$, there force the **scalar-vector multiplication is associative**.

#### Scalar multiplication is left distributitive

If the scalar-vector multiplication is left distributive, it means that we should be able to write the following equality :

$(x+y)\textbf{a} = x\textbf{a} + y\textbf{a}$

Let's prove it !

On the left side, we have a scalar addition and a preceeding scalar-vector multiplication

$(x+y)\textbf{a} = (x+y)\begin{bmatrix} a_{1} \\ ... \\ a_{d} \end{bmatrix} = \begin{bmatrix} (x+y)a_{1} \\ ... \\ (x+y)a_{d} \end{bmatrix} = \begin{bmatrix} xa_{1}+ya_{1} \\ ... \\ xa_{d}+ya_{d} \end{bmatrix}$

On the right side, we have vectors addition preceed by two scalar-vector multiplication

$x\textbf{a} + y\textbf{a} = x\begin{bmatrix} a_{1} \\ ... \\ a_{d} \end{bmatrix} + y\begin{bmatrix} a_{1} \\ ... \\ a_{d} \end{bmatrix} = \begin{bmatrix} xa_{1} \\ ... \\ xa_{d} \end{bmatrix} + \begin{bmatrix} ya_{1} \\ ... \\ ya_{d} \end{bmatrix} = \begin{bmatrix} xa_{1}+ya_{1} \\ ... \\ xa_{d}+ya_{d} \end{bmatrix}$

We prove the equality is correct, therefore we can confirm that **scalar-vector multiplication is left distributive**.

In addition, we can tell that **scalar-vector multiplication is right distributive**, as $\textbf{a}(x+y) = \textbf{a}x + \textbf{a}y$.

Then according to all the demonstration we've made we can tell that $x(\textbf{a}+\textbf{b}) = x\textbf{a} + x\textbf{b}$ will also be verified as it is another version of the right distributive property.

### For the following pairs of vectors find the angle they form and explain how you do it.

Angle between two nonzero vectors a, b defined as

$$\angle (\textbf{a},\textbf{b}) = \arccos \left(\frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{a}\rVert\lVert\textbf{b}\rVert}\right)$$

The *Euclidean norm* (or just norm) of an n-vector **x** is

$$ \lVert\textbf{x}\rVert = \sqrt{x_{1}^2 + \ldots + x_{n}^2} = \sqrt{x^T + x} $$

#### For $(\begin{bmatrix} 1 \\ 1 \end{bmatrix}, \begin{bmatrix} 1 \\ 2 \end{bmatrix})$

$\textbf{a}^T \cdot \textbf{b} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}^T \cdot \begin{bmatrix} 1 \\ 2 \end{bmatrix} = \begin{bmatrix} 1 & 1 \end{bmatrix} \cdot \begin{bmatrix} 1 \\ 2 \end{bmatrix} = (1 \times 1) + (1 \times 2) = 3$ 

$\lVert\textbf{a}\rVert = \sqrt{a^T + a} = \sqrt{a_{1}^2 + a_{2}^2} = \sqrt{1^2 + 1^2} = \sqrt{2} $

$\lVert\textbf{b}\rVert = \sqrt{b^T + b} = \sqrt{b_{1}^2 + b_{2}^2} = \sqrt{1^2 + 2^2} = \sqrt{5} $

$\angle (\textbf{a},\textbf{b}) = \arccos \left(\frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{a}\rVert\lVert\textbf{b}\rVert}\right) = \arccos \left(\frac{3}{\sqrt{2}\sqrt{5}}\right) \approx 18.43^\circ$

#### For $(\begin{bmatrix} 1 \\ 1 \end{bmatrix}, \begin{bmatrix} -1 \\ -1 \end{bmatrix})$

$\textbf{a}^T \cdot \textbf{b} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}^T \cdot \begin{bmatrix} -1 \\ -1 \end{bmatrix} = \begin{bmatrix} 1 & 1 \end{bmatrix} \cdot \begin{bmatrix} -1 \\ -1 \end{bmatrix} = (1 \times -1) + (1 \times -1) = -2$ 

$\lVert\textbf{a}\rVert = \sqrt{a^T + a} = \sqrt{a_{1}^2 + a_{2}^2} = \sqrt{1^2 + 1^2} = \sqrt{2} $

$\lVert\textbf{b}\rVert = \sqrt{b^T + b} = \sqrt{b_{1}^2 + b_{2}^2} = \sqrt{-1^2 + -1^2} = \sqrt{2} $

$\angle (\textbf{a},\textbf{b}) = \arccos \left(\frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{a}\rVert\lVert\textbf{b}\rVert}\right) = \arccos \left(\frac{-2}{\sqrt{2}\sqrt{2}}\right) = 180^\circ$

#### For $(\begin{bmatrix} 1 \\ 1 \end{bmatrix}, \begin{bmatrix} 2 \\ 2 \end{bmatrix})$

$\textbf{a}^T \cdot \textbf{b} = \begin{bmatrix} 1 \\ 1 \end{bmatrix}^T \cdot \begin{bmatrix} 2 \\ 2 \end{bmatrix} = \begin{bmatrix} 1 & 1 \end{bmatrix} \cdot \begin{bmatrix} 2 \\ 2 \end{bmatrix} = (1 \times 2) + (1 \times 2) = 4$ 

$\lVert\textbf{a}\rVert = \sqrt{a^T + a} = \sqrt{a_{1}^2 + a_{2}^2} = \sqrt{1^2 + 1^2} = \sqrt{2} $

$\lVert\textbf{b}\rVert = \sqrt{b^T + b} = \sqrt{b_{1}^2 + b_{2}^2} = \sqrt{2^2 + 2^2} = \sqrt{8} = \sqrt{2^2 \times 2} = 2\sqrt{2} $

$\angle (\textbf{a},\textbf{b}) = \arccos \left(\frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{a}\rVert\lVert\textbf{b}\rVert}\right) = \arccos \left(\frac{4}{2\sqrt{2}\sqrt{2}}\right) = 0^\circ$

### For each of the above vector pairs compute the distances of the involved vectors, visualise the vectors and explain to what does the distance correspond to.

*(Euclidean)* distance between n-vectors **a** and **b** is

$$\textbf{dist}\left(a, b\right) = \lVert a - b \lVert$$

```python
def compute_distance(a, bi):
    
    # Compute the distance between the two vectors
    distance = np.linalg.norm(a - bi)

    # Create a figure and axis object
    fig, ax = plt.subplots()
    ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='a')
    ax.quiver(0, 0, bi[0], bi[1], angles='xy', scale_units='xy', scale=1, color='green', label='b1')
    ax.plot([a[0], bi[0]], [a[1], bi[1]], linestyle='--', color='red')
    ax.annotate('distance = {:.2f}'.format(distance), xy=(a+bi)/2, xytext=(3, 4), textcoords='offset points')
    
    # Calculate the limits of the x and y axes
    minx = min(0, a[0], bi[0]) - 1
    maxx = max(0, a[0], bi[0]) + 1 
    ax.set_xticks(range(minx,maxx))
    ax.set_xlim([minx,maxx])
    miny = min(0, a[1], bi[1]) - 1
    maxy = max(0, a[1], bi[1]) + 1
    ax.set_yticks(range(miny,maxy))
    ax.set_ylim([miny,maxy])

    # Make the spines meet at (0,0)
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Add labels for the x and y axes
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    # Add a legend
    ax.legend()

    # return the plot
    return plt
```

For $\textbf{dist}(\begin{bmatrix} 1 \\ 1 \end{bmatrix}, \begin{bmatrix} 1 \\ 2 \end{bmatrix}) =  \lVert \begin{bmatrix} 1 \\ 1 \end{bmatrix} - \begin{bmatrix} 1 \\ 2 \end{bmatrix} \lVert = \lVert \begin{bmatrix} 0 \\ -1 \end{bmatrix} \lVert = \sqrt{0^2 + -1^2} = \sqrt{1} = 1$

```python
a = np.array([1, 1])
b1 = np.array([1, 2])

plt = compute_distance(a, b1)
plt.show()
```

For $\textbf{dist}(\begin{bmatrix} 1 \\ 1 \end{bmatrix}, \begin{bmatrix} -1 \\ -1 \end{bmatrix}) =  \lVert \begin{bmatrix} 1 \\ 1 \end{bmatrix} - \begin{bmatrix} -1 \\ -1 \end{bmatrix} \lVert = \lVert \begin{bmatrix} 2 \\ 2 \end{bmatrix} \lVert = \sqrt{2^2 + 2^2} = 2\sqrt{2}$

```python
b2 = np.array([-1, -1])

plt = compute_distance(a, b2)
plt.show()
```

For $\textbf{dist}(\begin{bmatrix} 1 \\ 1 \end{bmatrix}, \begin{bmatrix} 2 \\ 2 \end{bmatrix}) =  \lVert \begin{bmatrix} 1 \\ 1 \end{bmatrix} - \begin{bmatrix} 2 \\ 2 \end{bmatrix} \lVert = \lVert \begin{bmatrix} -1 \\ -1 \end{bmatrix} \lVert = \sqrt{-1^2 + -1^2} = \sqrt{2}$

```python
b3 = np.array([2, 2])

plt = compute_distance(a, b3)
plt.show()
```

### Show how can you produce a vector that has the same direction as $\begin{bmatrix}2 \\ 3\end{bmatrix}$ and a norm of 1.

To have a vector **v** with the same direction as the $\begin{bmatrix}2 \\ 3\end{bmatrix}$, we need to find a scalar $k > 0$ so that $\textbf{v} = k \times \begin{bmatrix}2 \\ 3\end{bmatrix}$.

$\lVert k \times \begin{bmatrix}2 \\ 3\end{bmatrix} \lVert = 1$

$\mid k \mid \times \lVert \begin{bmatrix}2 \\ 3\end{bmatrix} \lVert = 1$

$k \times \sqrt{2^2 + 3^2} = 1$

$k \times \sqrt{13} = 1$

$k = \frac{1}{\sqrt{13}}$

Which means that you get $k = \frac{1}{\lVert\begin{bmatrix}2 \\ 3\end{bmatrix}\lVert}$ and therefore we have 

$ \textbf{v} = k \times \begin{bmatrix}2 \\ 3\end{bmatrix}$

$ \textbf{v} = \frac{1}{\lVert\begin{bmatrix}2 \\ 3\end{bmatrix}\lVert} \times \begin{bmatrix}2 \\ 3\end{bmatrix}$

$ \textbf{v} = \frac{\begin{bmatrix}2 \\ 3\end{bmatrix}}{\lVert\begin{bmatrix}2 \\ 3\end{bmatrix}\lVert}$

We can then conclude that to get a vector with the same direction of another one but with a norm of one, you just have to divide this vector by his own norm.

**This is the definition of unit/normalized vector $\hat{x}=\frac{x}{\lVert x \lVert}$**


### Given the vectors $\textbf{a} = \begin{bmatrix}2 \\ 4\end{bmatrix}$ and $\textbf{b} = \begin{bmatrix}1 \\ 1\end{bmatrix}$, explain what is the projection of $\textbf{a}$ on $\textbf{b}$, show how to compute it, compute it, and visualise it

So if $\textbf{a}$ & $\textbf{b}$ are two vector that form an angle $\theta$, the projection *d* of **a** on **b** or what we call the *signed component* of **a** on **b** is the norm of the vector **d**. This vector has the same or opposite direction as **b** and obtained by projecting **a** "orthogonally" on **b**. 

Knowing that we can now try to calculate the $\lVert \textbf{d}\lVert$ and finally get **d**.


#### 1) Express $\lVert \textbf{d}\lVert$

The vector **d** is in our case on the adjacent side of our right-angle triangle that has the vector **a** on the hypotenus of this latter.

As per the classic trigonometry $cos(\theta)=\frac{adj}{hyp}$ which in our case with vectors means $cos(\theta)=\frac{\lVert \textbf{d}\lVert}{\lVert \textbf{a}\lVert}$

So we can say that $\lVert \textbf{d}\lVert = \lVert \textbf{a}\lVert cos(\theta)$.

We can now either 
- Fist calculate the angle $\theta$ with the known function using trigonometry and then multiply the norm of **a** to obtain the norm of **d**

OR

- Find an equivalent of $\lVert \textbf{a}\lVert cos(\theta)$ to avoid using angle in our calculus for our research of the vector **d**



The second solution will be :

$\textbf{a}^T \cdot \textbf{b} = \lVert\textbf{a}\rVert\lVert\textbf{b}\rVert\cos(\theta)$

$\frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{b}\rVert} = \frac{\lVert\textbf{a}\rVert\lVert\textbf{b}\rVert\cos(\theta)}{\lVert\textbf{b}\rVert}$

$\frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{b}\rVert} = \lVert\textbf{a}\rVert\cos(\theta)$

**So as $\frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{b}\rVert} = \lVert\textbf{a}\rVert\cos(\theta)$ we can conclude that $\lVert \textbf{d}\lVert = \frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{b}\rVert}$**

#### 2) Determine the unit vector $\hat{\textbf{b}}$

As we already know, $\hat{\textbf{b}} = \frac{\textbf{b}}{\lVert\textbf{b}\rVert}$ 

#### 3) Find the vector **d**

Now that we know that $\textbf{d}$ is colinear with $\textbf{b}$ and we know how to get $\hat{\textbf{b}}$ and $\lVert\textbf{d}\rVert$ we might be able to calculate **d**

$\textbf{d} = \hat{\textbf{b}} \times \lVert\textbf{d}\rVert$

$\textbf{d} = \frac{\textbf{b}}{\lVert\textbf{b}\rVert} \times \frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{b}\rVert} = \frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{b}\rVert^2}\textbf{b}$

#### What about our example

First we calculate what we need

$\textbf{a}^T \cdot \textbf{b} = (2 \times 1) + (4 \times 1) = 6$ 

$\lVert\textbf{b}\rVert = \sqrt{b^T + b} = \sqrt{b_{1}^2 + b_{2}^2} = \sqrt{1^2 + 1^2} = \sqrt{2} $

$\hat{\textbf{b}} = \frac{\textbf{b}}{\lVert\textbf{b}\rVert} = \frac{\begin{bmatrix}1 \\ 1\end{bmatrix}}{\sqrt{2}}$

Then we can compute *d*
$\lVert \textbf{d}\lVert = \frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{b}\rVert} = \frac{6}{\sqrt{2}} = 3\sqrt{2}$


Finally we can compute **d**

$\textbf{d} = \frac{\textbf{b}}{\lVert\textbf{b}\rVert} \times \frac{\textbf{a}^T \cdot \textbf{b}}{\lVert\textbf{b}\rVert}$

$\textbf{d} = \frac{\begin{bmatrix}1 \\ 1\end{bmatrix}}{\sqrt{2}} \times 3\sqrt{2}$

$\textbf{d} = \begin{bmatrix}1 \\ 1\end{bmatrix} \times \frac{1}{\sqrt{2}} \times 3\sqrt{2}$

$\textbf{d} = \begin{bmatrix}1 \\ 1\end{bmatrix} \times \frac{3\sqrt{2}}{\sqrt{2}} $

$\textbf{d} = \begin{bmatrix}1 \\ 1\end{bmatrix} \times 3$

$\textbf{d} = \begin{bmatrix}3 \\ 3\end{bmatrix} $

```python
# Define two vectors
a = np.array([2, 4])
b = np.array([1, 1])

# Compute the projection of a onto b
proj = (np.dot(a, b) / pow(np.linalg.norm(b),2)) * b

# Plot the vectors and their projection
fig, ax = plt.subplots()
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u')
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='v')
ax.quiver(0, 0, proj[0], proj[1], angles='xy', scale_units='xy', scale=1, color='green', label='proj')

# Calculate the limits of the x and y axes
ax.set_xticks(range(0,5))
ax.set_xlim([0, 5])

ax.set_yticks(range(0,5))
ax.set_ylim([0,5])

# Make the spines meet at (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Add labels for the x and y axes
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a legend
ax.legend()
```
